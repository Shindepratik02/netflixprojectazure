```python
checkpoint_location="abfss://silver@netflixstorageacct.dfs.core.windows.net/checkpoint"
```

```python
df=spark.readStream.format("cloudFiles")\
  .option("cloudFiles.format","csv")\
  .option("cloudFiles.schemaLocation",checkpoint_location)\
  .load("abfss://raw@netflixstorageacct.dfs.core.windows.net/")
```

```python
df.writeStream.option("checkpointLocation",checkpoint_location)\
    .trigger(availableNow=True)\
    .start("abfss://bronze@netflixstorageacct.dfs.core.windows.net/netflix_titles")
```
