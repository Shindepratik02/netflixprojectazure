```python
dbutils.widgets.text("sourcefolder", "netflix_directors")
dbutils.widgets.text("targetfolder", "netflix_directors")
```

```python
sourcefolder = dbutils.widgets.get("sourcefolder")
targetfolder = dbutils.widgets.get("targetfolder")
```

```python
df=spark.read.format("csv").option("header","true").option("inferSchema","true").load(f"abfss://bronze@netflixstorageacct.dfs.core.windows.net/{sourcefolder}")
```

```python
df.display()
```

```python
df.write.format("delta").mode("append").option("path",f"abfss://silver@netflixstorageacct.dfs.core.windows.net/{targetfolder}").save()
```
